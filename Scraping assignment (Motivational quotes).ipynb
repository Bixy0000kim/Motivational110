```python
def scrape(soup):
    author_name = [i.text.strip() for i in soup.find_all('span', class_ = 'authorOrTitle')]
    quote = [list(i)[0].strip() for i in soup.find_all('div', class_ = 'quoteText')]
    likes = [int(i.text.split()[0]) for i in soup.find_all('a', class_ = 'smallText')]
    tags = [list(i.text.split()[1:]) for i in soup.find_all('div', class_ = 'greyText smallText left')]
    return pd.DataFrame({'AuthorName': author_name, 'Quote': quote, 'Likes': likes, 'Tags': tags})
```

```python
df = scrape(soup)
df.head()
```
